```python
def normalize(data):
    min_max_scaler = preprocessing.MinMaxScaler()
    column_names = list(data)
    activity = data.pop("Activity")
    x = data.values
    x_scaled = min_max_scaler.fit_transform(x)
    normalized_data = pd.DataFrame(x_scaled)
    normalized_data.insert(0,"Activity",activity,True)
    data.insert(0,"Activity",activity,True)
    normalized_data.columns = column_names
    
    return normalized_data

```

```python
#data.groupby('Activity').mean().plas.hist(alpha=0.4)
data_n = normalize(data)

data_n
```
